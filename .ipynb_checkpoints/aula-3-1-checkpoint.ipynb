# Vamos ver como utilizar o python para realizar uma predição?

* Vamos pensar juntos em uma tarefa de classificação e procurar por um algoritmo para fazermos o primeiro teste?

Um exemplo muito simples e muito conhecido para quem está começando no mundo dos dados é o do Titanic. Nesse examplo, procuramos predizer quem irá sobrevier e quem irá morrer. 

```python
# linear algebra
import numpy as np 

# data processing
import pandas as pd 

# data visualization
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style

# Algorithms
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.metrics import accuracy_score


```

```python

X, y = load_iris(return_X_y=True)
```

```python
iris = load_iris()
data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
data1.head()
```

<img src='figures/iris-machinelearning.png'>

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf) ;
```

```python
y_pred = clf.predict(X)
```

```python
accuracy_score(y, y_pred, normalize=True)
```

o que estamos vendo aqui: 
Esse algoritmo busca pelas variáveis mais importantes e de acordo com a combinação delas, nos diz a qual classe um objeto pertence;
